```python
import pandas as pd
dfTime = pd.read_csv("SamplingTime1.csv", names=['percent',
                 '100K',
                 '500K',
                 '1M'])
```

```python
print(dfTime)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
#plt.plot(dfTime['percent'], dfTime['100K'], color='red')
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.plot(dfTime['percent'], dfTime['100K'], color='red', linestyle = '--')
plt.plot(dfTime['percent'], dfTime['500K'], color='blue',linestyle = '-.')
plt.plot(dfTime['percent'], dfTime['1M'], color='green', linestyle = ':')
#plt.plot(dfTime['percent'], dfTime['10M'], color='black', linestyle = '-')

plt.legend(prop={'size':15})

plt.ylabel('Running Time (in s)', fontsize=24)
plt.xlabel('Percentage of Sample Data (in %)', fontsize=24)
plt.show()
```
